```python
#import needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
%matplotlib inline
```

```python
#uncomment for coursera platform
#noaa = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')

#uncomment for working offline
noaa = pd.read_csv('weather-data.csv')
```

```python
#convert date column to datetime
noaa['Date'] = pd.to_datetime(noaa['Date'])
```

```python
#remove leap days
noaa = noaa[~((noaa['Date'].dt.month==2)& (noaa['Date'].dt.day==29))]
```

```python
#create month and date columns

noaa['Month'] = noaa['Date'].dt.month
noaa['Day'] = noaa['Date'].dt.day
```

```python
#convert Data_Value to Degrees C
noaa['Degrees C'] = noaa['Data_Value']/10

```

```python
#set Month, Day multi-index
noaa.set_index(['Month', 'Day'], inplace=True)

```

```python
#Sort df by Date
noaa.sort_values('Date', inplace=True)
```

```python
#divide into 04-2014 and 2015

df04_14 = noaa[noaa['Date'].dt.year !=2015]
df15 = noaa[noaa['Date'].dt.year == 2015]
```

```python
#divide into max and min
tmin414 = df04_14[df04_14['Element']=='TMIN']
tmax414 = df04_14[df04_14['Element']=='TMAX']

tmin15 = df15[df15['Element']=='TMIN']
tmax15 = df15[df15['Element']=='TMAX']
```

```python
#get min and max for each day across all years
min414=tmin414.groupby(tmin414.index).min()[['Degrees C','Date']]
max414=tmax414.groupby(tmax414.index).max()[['Degrees C','Date']]

min15 = tmin15.groupby(tmin15.index).min()[['Degrees C','Date']]
max15 = tmax15.groupby(tmax15.index).max()[['Degrees C','Date']]
```

```python


#get record low/high data


rec_lo_days= min15[min15['Degrees C']<min414['Degrees C']]
rec_hi_days = max15[max15['Degrees C']>max414['Degrees C']]
```

```python
#create a date array to match the dates in 2015 data
adj_date = np.arange('2015-01-01','2016-01-01', dtype='datetime64[D]' )
```

```python
#add adjusted date column to 04-14 data
min414['Adj_Date'] = adj_date
max414['Adj_Date'] = adj_date
```

```python
#reset the index to adjusted dates
min414.set_index('Adj_Date', inplace=True)
max414.set_index('Adj_Date', inplace=True)
```

```python
#set up the canvas and choose the figure size
fig, ax = plt.subplots(figsize = (16, 8))

#line plot for Max and Min 04-14
ax.plot(min414.index, min414['Degrees C'], color='blue', alpha=0.37)
ax.plot(max414.index, max414['Degrees C'], color='red', alpha=0.37)
#plot record high and low of 2015
ax.plot(rec_lo_days['Date'], rec_lo_days['Degrees C'], 'v', color = 'blue')
ax.plot(rec_hi_days['Date'], rec_hi_days['Degrees C'], '^', color = 'red')
#shade area between lines
ax.fill_between(min414.index,min414['Degrees C'] ,max414['Degrees C'], color='orange', alpha = 0.37)
#change x axis to only include 1/1-12/31 
ax.set_xlim(datetime(2015, 1,1), datetime(2015,12,31))
#add legend
ax.legend(['2004-2014 Minimum Temperature', '2004-2014 Maximum Temperature', '2015 Record Low', '2015 Record High'])

#add title
ax.set_title('Weather Patterns near Ann Arbor, MI 2004-2015', fontsize=15)
#add x and y labels
ax.set_xlabel('Time of Year', fontsize=13)
ax.set_ylabel('Temperature in $^\circ$C', fontsize=13)
#set background color

#ax.set_facecolor('#F0ECDC')
#change date format to first three letters of the month
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b')) 
ax.tick_params(axis='x', labelsize=11)
#display the graph for none jupyter platforms
plt.savefig('weather_patterns.png')
plt.show()
```
